```python
import time
import random

lanes_group = [
    ["North", "South"],
    ["East", "West"]
]

class SmartTrafficSystem:
    def __init__(self):
        self.traffic = {}
        self.pedestrian_waiting = False
        self.last_green_group = None

    def update_traffic(self):
        for group in lanes_group:
            for lane in group:
                self.traffic[lane] = random.randint(0, 20)

    def get_priority_group(self):
        g0 = sum(self.traffic[l] for l in lanes_group[0])
        g1 = sum(self.traffic[l] for l in lanes_group[1])

        preferred = 0 if g0 >= g1 else 1


        if preferred == self.last_green_group:
            preferred = 1 - preferred

        return preferred

    def get_timing(self, total):
        if total >= 25:
            return "High Traffic", 60
        elif total >= 10:
            return "Normal Traffic", 30
        else:
            return "Low Traffic", 15

    def display(self, cycle, green_group, state, timer, group_index):
        print(f"CYCLE {cycle}")

        print("\nTraffic Density:")
        for lane in ["North", "South", "East", "West"]:
            count = self.traffic[lane]
            bar = "█" * (count // 2)
            signal = "GREEN" if lane in green_group else "RED"
            print(f"{lane:<6}: {count:2d} cars | {bar:<10} | {signal}")

        print(f"\nMode: {state}")
        print("Pedestrian:", "WAITING" if self.pedestrian_waiting else "None")

        print(f"\nGREEN SIGNAL: {green_group}")
        print(" Green light countdown:")

        threshold = 8
        min_green = 15

        elapsed = 0

        while elapsed < timer:
            remaining = timer - elapsed
            print(f"   {remaining} sec remaining...")


            if elapsed >= min_green:
                current = sum(self.traffic[l] for l in lanes_group[group_index])
                opposite = sum(self.traffic[l] for l in lanes_group[1 - group_index])

                if opposite > current + threshold:
                    print("   Heavy traffic on opposite → switching early!")
                    break

            time.sleep(1)
            elapsed += 1

        print("   Green phase ended!")


        print("\nSwitching to YELLOW (current lanes prepare to stop)...")
        for t in range(3, 0, -1):
            print(f"   {t} sec remaining (YELLOW)...")
            time.sleep(1)

        print("   Yellow phase ended!")


        print("\nALL LANES YELLOW (prepare to switch)...")
        for t in range(2, 0, -1):
            print(f"   {t} sec remaining (ALL-YELLOW)...")
            time.sleep(1)

        print("   Switching sides now!")

    def run(self, cycles=5):
        for i in range(1, cycles + 1):

            self.update_traffic()
            self.pedestrian_waiting = random.choice([True, False])

            group_index = self.get_priority_group()
            green_group = lanes_group[group_index]

            total = sum(self.traffic[l] for l in green_group)
            state, timer = self.get_timing(total)

            if self.pedestrian_waiting:
                state = "Pedestrian Priority"
                timer = 15

            self.display(i, green_group, state, timer, group_index)

            self.last_green_group = group_index



system = SmartTrafficSystem()
system.run(5)
```
